```python
# Simple_Linear_Regression_GD_method_14.ipynb

# Cell No. 1

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
from sklearn.linear_model import LinearRegression

# Input feature: House areas in throusand square feet
X = np.array([1.4, 1.8, 2.0, 1.6, 2.4, 1.2, 2.6, 2.8, 2.2, 3.2, 3.0, 3.4])
m = len(X)

# Target feature: House prices in thousands of currency units
Y = np.array([220,300,350,250,450,200,470,490,430,500,520,550]) 

# Create a scatter plot of the data
plt.scatter(X, Y, marker='x', c='r') 
# Set the plot title and the labels of the axes
plt.title("House selling Price Prediction")
plt.ylabel('House areas in throusand square feet ')
plt.xlabel('House prices in thousands of currency units ')
plt.show()
```

```python
# Cell No. 2
# Training of the model parameters 

# Defining the cost function (averaged residual)
def MSE(w, b, X, Y):
    return 1/(len(Y))*np.sum(((w*X + b) - Y)**2)

# Define functions to calculate partial derivatives

def dEdm(w, b, X, Y):
    res = 2/len(X)*np.dot(w*X + b - Y, X)
    return res

def dEdb(w, b, X, Y):
    res = 2/len(X)*np.sum(w*X + b - Y)
    return res

# Defining function for the gradient decent

def gradient_descent(dEdm, dEdb, w, b, X, Y, learning_rate, maxiter):
    for iteration in range(maxiter):
        w = w - learning_rate*dEdm(w, b, X, Y)
        b = b - learning_rate*dEdb(w, b, X, Y)

        cost = MSE(w, b, X, Y)

        if cost < 0.01:
            break

    #print ("Cost after iteration", iteration, " is ", cost)

    return w, b

# Model building using gradient decent with learning rate = 0.01
# with parameter initialization  w=100, b=5, maximum iterations=1000

w, b = gradient_descent(dEdm, dEdb, 100.0, 5.0, X, Y, 0.01, 1000)
```

```python
# Cell No. 3

print("Slope of the regression line:", w)
print("y-Intercept of the regression line:", b)

# predicting the values on the given x_values with the trained model

Y_pred = w * X + b
```

```python
# Cell No. 4
# plot the original data points and the regression line

# reshaping for plotting (effective when needed)
Y_pred = Y_pred.reshape(-1)  

# Scatter plot of the actual data
plt.scatter(X, Y, color='blue', label='Actual Data')
# Regression line
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.xlabel('X') ; plt.ylabel('Y') ; plt.legend() ; plt.show()
```
