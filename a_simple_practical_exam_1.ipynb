```python
# Q1(i)	Create a Data Frame (DF) from a dictionary that have exactly 
# 6 instances/rows, then use index attribute to re-label its indices.
# Use your own data values. Don’t set its name/identifier as df
# Set some other name/identifier of the DF in your code.

import numpy as np
import pandas as pd

grades_dict = {'Waqas':[87,96,np.nan,80,90,75],\
               'Ejaz':[100,80,90,np.nan,np.nan,np.nan],\
               'Saad':[94,89,np.nan,np.nan, 98,90],\
             'Kinza':[100,93,83,94,81,82],\
               'Bano':[83,65,100,90,30,85]}

grades=pd.DataFrame(grades_dict)
grades           
```

```python
grades.index=['Test1','Test2','Test3','Test4','Test5','Test6']
grades
```

```python
# Q1(ii) Write two different statements/commands, each of which 
# can access any specific column of the DF created in (i)

grades['Bano']
```

```python
grades.Bano
```

```python
# Q1(iii) Write statements to explain  the use of loc and iloc attributes 
# for slicing of the DF created in (i)

grades.loc['Test2':'Test4']
```

```python
# Q1(iv) Write a command to sort the DF by column Indices in descending 
# order. Do not create a new sorted copy; modify the original DF.

grades.sort_index(axis=1, ascending=False, inplace=True)
grades
```

```python
# Q1(v) Write a command to remove all those columns of the new sorted DF
# that have more than two np.nan values.Do not create a new sorted copy;
# modify the data frame under consideration.

grades.dropna(axis=1, thresh=4, inplace=True)
grades
```

```python
# Q2(i) Assume that Amjad, Usman, Asif, Shahzad, and Mudassar are teachers 
# at a department. Their room numbers are 111, 103, 108, 110, and 109,
# respectively. Create a Data Frame (TA) of two columns to handle this data.
# Label the columns appropriately.

TA = pd.DataFrame({'Teacher':['Amjad','Usman','Asif',\
                             'Shahzad', 'Mudassar'],\
                  'Room':  [111,103,108,110,109]})
TA
```

```python
# Q2(ii) Assume that Amjad, Shahzad, and Mudassar have expertise in 
# “Applied Math”, whereas Usman and Asif have expertise in “Pure Math”.
# Create a variable that defines mapping of each teacher to his area 
# of expertise.

teacher_to_area = {'Amjad':'Applied Math', 'Usman':'Pure Math',\
                             'Asif':'Pure Math','Shahzad':'Applied Math',\
                   'Mudassar':'Applied Math'}

teacher_to_area 
```

```python
# Q2(iii) Add another column to TA, with label “Area”, through mapping so
# that it shows the name of the area of expertise in the row of each teacher

TA['Area'] = TA['Teacher'].map(teacher_to_area )
TA
```
