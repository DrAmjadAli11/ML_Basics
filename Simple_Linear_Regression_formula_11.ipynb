```python
# Simple_Linear_Regression_formula_11.ipynb

# Cell No. 1

import numpy as np
import matplotlib.pyplot as plt

# Input feature: House areas in throusand square feet
X = np.array([1.4, 1.8, 2.0, 1.6, 2.4, 1.2, 2.6, 2.8, 2.2, 3.2, 3.0, 3.4])
m = len(X)

# Target feature: House prices in thousands of currency units
Y = np.array([220,300,350,250,450,200,470,490,430,500,520,550]) 

# Create a scatter plot of the data
plt.scatter(X, Y, marker='x', c='r') 

# Set the title
plt.title("House selling Price Prediction")
# Set the y-axis label
plt.ylabel('House areas in throusand square feet ')
# Set the x-axis label
plt.xlabel('House prices in thousands of currency units ')
plt.show()
```

```python
# Cell No. 2

# Defining a function fit_linear_regression for parameters calculation
def fit_linear_regression(X,Y):
   
    w = ( m*np.sum(X*Y) -np.sum(X) * np.sum(Y) ) / \
                 ( m*np.sum(X*X) - np.sum(X)*np.sum(X) )
    b = np.mean(Y) - w * np.mean(X)
    
    return w, b

# Calling the function for model parameters using the training data
w, b = fit_linear_regression(X,Y)
```

```python
# Cell No. 3

print("Slope of the regression line:", w)
print("y-Intercept of the regression line:", b)

# predicting the values on the given x_values with the trained model

Y_pred = w*X + b
```

```python
# Cell No. 4
# plot the original data points and the regression line

# reshaping for plotting (effective when needed)
Y_pred = Y_pred.reshape(-1)  

# Scatter plot of the actual data
plt.scatter(X, Y, color='blue', label='Actual Data')
# Regression line
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.xlabel('X') ; plt.ylabel('Y') ; plt.legend() ; plt.show()
```
